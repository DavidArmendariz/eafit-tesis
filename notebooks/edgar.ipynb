```python
import os

import pandas as pd
import sec_api
from dotenv import load_dotenv
from IPython.display import HTML
```

```python
load_dotenv()
```

```python
fullTextSearchApi = sec_api.FullTextSearchApi(api_key=os.getenv("SEC_API"))

query = {
    "query": "lease agreement between -amendment -addendum -extension",
    "formTypes": ["10-K"],
    "startDate": "2020-01-01",
    "endDate": "2023-12-31",
    "page": 2,
}

filings = fullTextSearchApi.get_filings(query)

print(f"Total filings found: {filings['total']['value']}")
```

```python
filings_data = filings["filings"]
```

```python
df = pd.DataFrame(filings_data)
df.columns
```

```python
df = df[["companyNameLong", "description", "filingUrl"]]
df = df[~df["description"].isin(["10-K", "FORM 10-K"])]
```

```python
pd.set_option("display.max_rows", None)

HTML(df.to_html(render_links=True, escape=False))
```
